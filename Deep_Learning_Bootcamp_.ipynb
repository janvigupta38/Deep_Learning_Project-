Importing Libraries

```python
import tensorflow as tf
from  tensorflow.keras.models import Sequential
from  tensorflow.keras.layers import Input, Dense
import numpy as np
```

Creating Dataset for an AND Gate

```python
x=np.array([[0,0],[0,1],[1,0],[1,1]],"uint8")
y=np.array([[0],[0],[0],[1]],"uint8") #uint- in numpy it stored no as 8 byte
```

Creating a Sequential Model using tensorflow

```python
model=Sequential() 
model.add(Dense(units=1,activation='sigmoid',input_dim=2)) # o/p is only one
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
model.summary()
#Dense-Every neuron of this layer is connected to previous layer
#input_dim=2- two inputs
#adam - for weight updation, 
#binay_crossentropy-when o/p is 0 or 1 , 
#metrics-we need accuracy of model
```

```python
model.fit(x,y,epochs=500)
```

```python
print(model.layers[0].get_weights()) #0 is the first layer
```

```python
model.predict(x) #this is for sigmoid function threshold is 0.5
#value>0.5- +ve result otherwise -ve result
```

Creating Database for A XOR Gate

```python
x=[[0,0],[0,1],[1,0],[1,1]]
y=[[0],[1],[1],[0]]
```

```python
xor=Sequential() 
xor.add(Dense(2,input_dim=2,activation='sigmoid')) # hidden layer with 2 neurons
xor.add(Dense(1,activation='sigmoid',)) #o/p= 1 neuron
xor.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
xor.summary()
```

```python
xor.fit(x,y,epochs=1500)
```

```python
print(xor.layers[0].get_weights()) #0 is the first layer
```

```python
xor.predict(x)
```

Deep Neural Network to Predict hand writing Digit Recognition

```python
from sklearn.datasets import load_breast_cancer
Data=load_breast_cancer()
```

```python
print(Data.DESCR)
```

```python
features=Data.data
target=Data.target

print(features.shape)
print(target.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.2)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential() 
model.add(Dense(32,input_dim=30,activation='relu')) # hidden layer with 2 neurons
model.add(Dense(64,activation='relu'))# hidden layer 2
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train,y_train,epochs=200,batch_size=32)
```

```python
score=model.evaluate(X_test,y_test)
print(score)
```

```python
predictions=model.predict(X_test)
label=[]
for pred in predictions:
  if pred>=0.5:
    print("Malignent")
  else:
    print("Benign")
```
